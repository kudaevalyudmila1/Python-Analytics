### Задание 1.
#### Считать данные с помощью pandas. Вывести на экран первые 5 строк

```python
import pandas as pd
df = pd.read_csv('./kc_house_data.csv', sep=',')
df.head()
```

#### 1.2 Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома. Если реновации дома не было, то в новом признаке поставьте 0.

```python

df['delta_renovated'] =  df['yr_renovated']- df['yr_built']
df['delta_renovated'] = df['delta_renovated'].apply(lambda x: 0 if x<0 else x)
df.head()
```

#### 1.3 Создайте признаки года продажи, месяца продажи

```python
df['year_sale'] = df['date'].apply(lambda x: x[:4])
df.head()
```

```python
df['month_sale'] = df['date'].apply(lambda x: x[4:6])
df.head()
```

#### 1.4 Удалите признаки date, zipcode, lat, long

```python
df1 = df.copy()
df1 = df1.drop(columns=['date','zipcode', 'lat', 'long'])
df1.head()
```
