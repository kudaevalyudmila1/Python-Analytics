```python
import pandas as pd
df = pd.read_csv('./laptops_10.csv', sep=',')
df.head(1)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(6,4))
plt.hist(df['Price_euros'])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Example')
plt.xticks(rotation=30)
```

```python
plt.figure(figsize=(6,4))
plt.hist(df['Memory_Amount'])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Memory')
plt.xticks(rotation=30);
```

```python
plt.figure(figsize=(6,4))
plt.hist(df['Weight'])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Count')
plt.xticks(rotation=30);
```

```python
data = df['Memory_Type'].value_counts()

plt.pie(data.values, autopct='%1.2f%%',labels=data.index)

plt.legend('Count')
plt.title('Memory_Type')
plt.xticks(rotation=30);
```

```python
data1 = df['Company'].value_counts()

plt.pie(data1.values, autopct='%1.2f%%',labels=data1.index)

plt.legend('Company')
plt.title('Memory_Type')
plt.xticks(rotation=30);
```

```python

plt.bar(data1.index, data1.values)
plt.xticks(rotation=90);
```

```python
data2 = df['OpSys'].value_counts()
plt.bar(data2.index, data2.values)
plt.xticks(rotation=90);
```

```python
data3 = df['Cpu_Company'].value_counts()

plt.pie(data3.values, autopct='%1.2f%%',labels=data3.index)

plt.legend('Cpu_Company')
plt.title('Cpu_Company')
plt.xticks(rotation=90);
```
